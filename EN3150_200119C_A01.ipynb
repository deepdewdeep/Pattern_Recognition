```python
def generate_signal(signal_length, num_nonzero):
    signal = np.zeros(signal_length)
    nonzero_indices = np.random.choice(signal_length, num_nonzero, replace=False)
    nonzero_values = 50 * np.random.randn(num_nonzero)
    signal[nonzero_indices] = nonzero_values
    return signal

# Data generation
signal_length = 100  # Total length of the signal
num_nonzero = 10
your_index_no = 200119  #index number (without English letter and without leading zeros)
signal = generate_signal(signal_length, num_nonzero)
signal[10] = (your_index_no % 10) * 10 + 10
if your_index_no % 10 == 0:
    signal[10] = np.random.randn(1) + 30
signal = signal.reshape(signal_length, 1)

# plotting
plt.figure(figsize=(15, 5))
plt.subplot(1, 1, 1)
plt.title("Data")
plt.stem(signal)
plt.show()
```

## MaxAbsScaler normalization

```python
# MaxAbsScaler normalization
scaler = MaxAbsScaler()
normalized_signal = scaler.fit_transform(signal)

# Plotting original and normalized signals
plt.figure(figsize=(15, 5))

# Original signal
plt.subplot(2, 1, 1)
plt.title("Original Data")
plt.stem(signal)

# Normalized signal
plt.subplot(2, 1, 2)
plt.title("Normalized Data (MaxAbsScaler)")
plt.stem(normalized_signal)

plt.tight_layout()
plt.show()
```

```python
# Min-Max scaling
def min_max_scale(data):
    min_val = np.min(data)
    max_val = np.max(data)
    scaled_data = (data - min_val) / (max_val - min_val)
    return scaled_data

# Standardization scaling (Z-score normalization)
def standardize_scale(data):
    mean_val = np.mean(data)
    std_val = np.std(data)
    scaled_data = (data - mean_val) / std_val
    return scaled_data

# Apply Min-Max scaling and Standardization scaling
min_max_scaled_signal = min_max_scale(signal)
standardized_signal = standardize_scale(signal)

# Plotting original, Min-Max scaled, and standardized signals
plt.figure(figsize=(15, 10))

# Original signal
plt.subplot(3, 1, 1)
plt.title("Original Data")
plt.stem(signal)

# Min-Max scaled signal
plt.subplot(3, 1, 2)
plt.title("Min-Max Scaled Data")
plt.stem(min_max_scaled_signal)

# Standardized signal
plt.subplot(3, 1, 3)
plt.title("Standardized Data")
plt.stem(standardized_signal)

plt.tight_layout()
plt.show()
```

```python
# Count non-zero elements in the original data
count_nonzero_original = np.count_nonzero(signal)

# Count non-zero elements in the Min-Max scaled data
count_nonzero_min_max = np.count_nonzero(min_max_scaled_signal)

# Count non-zero elements in the standardized data
count_nonzero_standardized = np.count_nonzero(standardized_signal)

# Print the counts
print("Number of non-zero elements in the original data:", count_nonzero_original)
print("Number of non-zero elements in the Min-Max scaled data:", count_nonzero_min_max)
print("Number of non-zero elements in the standardized data:", count_nonzero_standardized)
```

## analyzing the scaling algorithms

```python
# Plot histograms for original, Min-Max scaled, and standardized data
plt.figure(figsize=(15, 5))

# Original data
plt.subplot(1, 3, 1)
plt.hist(signal, bins=20, color='blue', alpha=0.7)
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Min-Max scaled data
plt.subplot(1, 3, 2)
plt.hist(min_max_scaled_signal, bins=20, color='green', alpha=0.7)
plt.title("Min-Max Scaled Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Standardized data
plt.subplot(1, 3, 3)
plt.hist(standardized_signal, bins=20, color='red', alpha=0.7)
plt.title("Standardized Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
